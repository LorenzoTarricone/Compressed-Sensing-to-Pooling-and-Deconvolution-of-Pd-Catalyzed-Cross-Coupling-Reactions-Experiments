```python

def visualize_plate_from_dict(df_dict):
  """
  Visualize a plate plotting it as an heatmap. (it shouldn't have nans)

  Args:
    df: A dictionary of pandas dataframe with the plate data indexed by the sample number

  Returns:
    A matplotlib figure with the plates
  """
  fig, axs = plt.subplots(2, 1, figsize=(10, 6))
  cax0 = axs[0].imshow(df_dict[1], aspect='auto')
  cax1 = axs[1].imshow(df_dict[2], aspect='auto')

  # Add colorbars
  fig.colorbar(cax0, ax=axs[0])
  fig.colorbar(cax1, ax=axs[1])

  # Add titles
  axs[0].set_title('Plate 1')
  axs[1].set_title('Plate 2')

  # Add x and y labels
  axs[0].set_ylabel('Plate Row')
  axs[1].set_ylabel('Plate Row')

  # Add tick labels as catalysts names
  axs[1].set_xticks(np.arange(len(df_dict[1].columns)))
  axs[1].set_xticklabels(df_dict[1].columns, rotation=90)


  plt.show()
```

```python
def visualize_plate_single(array, plate_number, time_point):
  """
  Visualize a plate plotting it as an heatmap.

  Args:
    array: An np array of shape (8,12)

  Returns:
    A matplotlib figure with the plates

    """

  array = array[:, :, plate_number, time_point]
  fig, ax = plt.subplots(figsize=(10, 6))
  cax = ax.imshow(array, aspect='auto')

  # Add colorbar
  fig.colorbar(cax, ax=ax)

  # Add titles
  ax.set_title('Plate ' + str(plate_number + 1) + ' at time point ' + str(time_point + 1))

  # Add x and y labels
  ax.set_ylabel('Solvents/bases')
  ax.set_xlabel('Catalysts')

  # Add tick labels as catalysts names
  if plate_number != 4:
    names = ['meCgPPh','Piperidinyl-amino-pincer', 'TrixiePhos','IPENT Cl','RuPhos', 'N-XantPhos', 'RockPhos','DiMel-HeptCl','VPhos', 'PPh_3',  'tBuBrettPhos', 'SIPr','CPhos',  'DPEPhos',  'AlPhos','Triisobutyl-Phosphatrane','SPhos', 'XantPhos','Me3(OMe)-tBuXPhos','Amphos','Jackie CPhos' ,  'dppp', 'AdBrettPhos', 'Mordal-Phos', 'Jackie Phos', 'dppf', 'AdBippyPhos','P(tBu)_3','GPhos',    'BINAP','BippyPhos','cataCXlum A','BrettPhos', 'dppdtbpf','QPhos', 'cBridP','CyJohnPhos', 'dtbpf','JosiPhosSL J009-1','PCy_3', 'XPhos','tBu-Phosphonito','tBuXPhos', 'nothing', 'nothing', 'nothing', 'nothing', 'nothing']
    ax.set_xticks(np.arange(array.shape[1]))
    ax.set_xticklabels(names[(plate_number)*12:(plate_number+1)*12], rotation=90)
  if plate_number == 4:
    names = ['mixture 1', 'mixture 2', 'mixture 3', 'mixture 4', 'mixture 5', 'mixture 6', 'mixture 7',  'mixture 8', 'mixture 9', 'mixture 10', 'mixture 11', 'mixture 12'  ]
    ax.set_xticks(np.arange(array.shape[1]))
    ax.set_xticklabels(names, rotation=90)

  #write the values of the array inside the array itslelf in white
  for i in range(len(array)):
    for j in range(len(array[0])):
      ax.text(j, i, round(array[i, j], 2), ha="center", va="center", color="white")


  plt.show()
```
