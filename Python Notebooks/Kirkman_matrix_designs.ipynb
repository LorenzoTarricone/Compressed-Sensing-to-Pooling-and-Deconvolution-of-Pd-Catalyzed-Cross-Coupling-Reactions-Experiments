```python
import numpy as np
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
```

## 9 Experiments design

```python
# Given triplets for Parallel Classes
parallel_triplets = [
    [1, 2, 3], [4, 5, 6], [7, 8, 9],  # Parallel Class 1
    [1, 4, 7], [2, 5, 8], [3, 6, 9],  # Parallel Class 2
    [1, 5, 9], [2, 6, 7], [3, 4, 8]   # Parallel Class 3
]

# Initialize an array of 0s, with 9 rows (one for each number) and 9 columns (one for each triplet)
parallel_matrix = np.zeros((9, 9), dtype=int)

# Fill the matrix with 1s according to the triplets
for col_idx, triplet in enumerate(parallel_triplets):
    for number in triplet:
        parallel_matrix[number-1, col_idx] = 1  # -1 to adjust for 0-based indexing

parallel_matrix
```

```python
# Plot the data with a specified size
fig, ax = plt.subplots(figsize=(15, 8))
cax = ax.matshow(parallel_matrix, cmap='gray_r')

# Create a custom legend
black_patch = mpatches.Patch(color='black', label='1')
white_patch = mpatches.Patch(color='white', label='0')

# Place the legend outside the plot
ax.legend(handles=[black_patch, white_patch], loc='upper center', bbox_to_anchor=(1.04, 1))
plt.title(f"Pooling matix M for pooling {parallel_matrix.shape[1]} catalysts in {parallel_matrix.shape[0]} experiments (assuming at most two hits)")
plt.show()
```

## 21 Experiments design
